# Circuits

### Question 1

Ans: 3

```python
qc1 = QuantumCircuit(1)
# Apply X and H-gate to the first:
qc1.x(0)
qc1.h(0)
qc1.draw('mpl')
```

```python
backend = BasicAer.get_backend('statevector_simulator')
job1 = execute(qc1, backend).result()
plot_bloch_multivector(job1.get_statevector(qc1))
```

### Question 2

Ans: 2

```python
qc2 = QuantumCircuit(2,2)
# Apply H-gate to the first:
qc2.h(0)
# Apply a CNOT:
qc2.cx(0,1)
qc2.barrier()
qc2.measure([0, 1], [0, 1])
qc2.draw('mpl')
```

```python
simulator = Aer.get_backend('qasm_simulator')
result=execute(qc2, backend=simulator).result()

plot_histogram(result.get_counts(qc2))
```

### Question 3

Ans: 1

```python
qc3 = QuantumCircuit(3)
# Apply H-gate to the first:
qc3.h(0)
qc3.h(1)

# Apply a CNOT:
qc3.cx(0,1)
qc3.cx(0,2)
qc3.draw('mpl')
```

```python
# Let's see the result:
backend = BasicAer.get_backend('statevector_simulator')
job = execute(qc3, backend).result()
plot_state_city(job.get_statevector(qc3), color=['midnightblue', 'midnightblue'],
     title="New State City")
```

### Question 4

Ans: 2

```python
from qiskit.quantum_info import Statevector
sv1 = Statevector.from_label('00')
plot_state_qsphere(sv1.data)
```

```python
mycircuit1 = QuantumCircuit(2)
mycircuit1.h(0)
#mycircuit1.x(1)
mycircuit1.cx(0,1)
mycircuit1.draw('mpl')
```

```python
new_sv1 = sv1.evolve(mycircuit1)
print(new_sv1)
plot_state_qsphere(new_sv1)
```

```python
backend = BasicAer.get_backend('statevector_simulator') # the device to run on
result = execute(mycircuit1, backend).result()
final_state  = result.get_statevector(mycircuit1)
```

```python
plot_state_city(final_state)
```

```python
plot_state_hinton(final_state)
```
