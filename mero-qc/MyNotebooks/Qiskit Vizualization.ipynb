```python
from qiskit import *
from qiskit.visualization import plot_histogram
from qiskit.tools.monitor import job_monitor
```

```python
#As a example we make 2 qubit Bell state
# quantum circuit to make a Bell state
bell = QuantumCircuit(2, 2)
bell.h(0)
bell.cx(0, 1)

meas = QuantumCircuit(2, 2)
meas.measure([0,1], [0,1])
```

```python
# execute the quantum circuit
backend = BasicAer.get_backend('qasm_simulator') # the device to run on
circ = bell+meas
result = execute(circ, backend, shots=1000).result()
counts  = result.get_counts(circ)
print(counts)
```

```python
plot_histogram(counts)
```

```python
# Execute 2-qubit Bell state again
second_result = execute(circ, backend, shots=1000).result()
second_counts  = second_result.get_counts(circ)
# Plot results with legend
legend = ['First execution', 'Second execution']
plot_histogram([counts, second_counts], legend=legend)
```

```python
# Execute 2-qubit Bell state again
third_result = execute(circ, backend, shots=1000).result()
third_counts  = second_result.get_counts(circ)
# Plot results with legend
legend = ['First execution', 'Second execution', 'Third Execution']
plot_histogram([counts, second_counts, third_counts], legend=legend)
```

```python
plot_histogram([counts, second_counts, third_counts], legend=legend, sort='desc', figsize=(15,12),
               color=['orange', 'black', 'red'], bar_labels=False)
```
